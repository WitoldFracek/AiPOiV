Proste progowanie obrazu na sk≈Çadowej V, modelu HSV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
img = cv2.imread('../../Dane1-4/Lena.jpg',cv2.IMREAD_COLOR)

img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

Image.fromarray(img_RGB)
```

```python
# zmiana obarzu na obraz w skali szaro≈õci
img_Gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

Image.fromarray(img_Gray)
```

```python
_ = plt.hist(img_Gray.flatten(), bins=255)
```

Dokumentacjƒô funkcji progujƒÖcej OpenCV znajdziesz pod linkiem: https://docs.opencv.org/4.5.3/d7/d1b/group__imgproc__misc.html#gae8a4a146d1ca78c626a53577199e9c57

Zwr√≥ƒá uwagƒô na r√≥≈ºne dostƒôpne typy progowania (`Threshold Types`)!

```python
# progowanie obrazu - metoda threshold OpenCV; parametry: obraz, pr√≥g, 
# w przypadku cv2.THRESH_BINARY_INV warto≈õci powy≈ºej progu uzyskajƒÖ warto≈õƒá trzeciego parametru (255)
(T, img_threshInv) = cv2.threshold(img_Gray, 170, 255, cv2.THRESH_BINARY_INV)

print('Pr√≥g: ',T)

Image.fromarray(img_threshInv)
```

```python
imgs = []
for i in range(255):
    (T, img_threshInv) = cv2.threshold(img_Gray, i, 255, cv2.THRESH_BINARY_INV)
    imgs.append(Image.fromarray(img_threshInv))
```

```python
imgs[0].save('thresholding.gif', save_all=True, append_images=imgs[1:])
```
