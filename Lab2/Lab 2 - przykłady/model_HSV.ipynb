Przykład rozjaśnienia obrazu poprzez działanie na składowej V, modelu barw HSV

```python
import cv2
import numpy as np

from PIL import Image
```

```python
# aby wczytać obraz
img = cv2.imread('../../Dane1-4/Lena.jpg')

# OpenCV domyślnie korzysta z modelu BGR (nie RGB!)
# aby wyświetlić, należy skonwertować:
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Image.fromarray(img_RGB)
```

Dokumentacja modułu konwersji modeli barw: https://docs.opencv.org/4.5.3/d8/d01/group__imgproc__color__conversions.html

Dokumentacja funkcji `cv2.add`: https://docs.opencv.org/4.5.3/d2/de8/group__core__array.html#ga10ac1bfb180e2cfda1701d06c24fdbd6

```python
# transformacja modelu barw z RGB do HSV
img_hsv = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)
height, width = img_hsv.shape[:2]

# rozbicie na składowe modelu HSV
h,s,v = cv2.split(img_hsv)

# wartość dodawana do składowej jasności
value = 50

v = cv2.add(v, value) # dlaczego ta operacja mogłaby być problematyczna, i dlaczego nie jest? podpowiedź: v.dtype

# scalenie składowych i wygenerowanie obrazu
img_hsv = cv2.merge([h, s, v])

# w celu wyświetlenia przechodzimy do RGB
img_output = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)
Image.fromarray(img_output)
```
